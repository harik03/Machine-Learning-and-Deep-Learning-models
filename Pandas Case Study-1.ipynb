```python
#6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 
```

```python
def value_segment(cust):
    if (cust['Customer_Value'] > 25000):
        return 'High Value Segment'
    elif (cust['Customer_Value'] > 10000) & (cust['Customer_Value'] < 25000):
        return 'Medium Value Segment'
    elif (cust['Customer_Value'] <= 10000):
        return 'Low Value Segment'
    
cust['Customer_Value_segment'] = cust.apply(value_segment, axis = 1)
cust
```

```python
#8. How many days between last purchase date and today?
```

```python
import datetime as dt

now = dt.datetime.today().strftime('%Y%m%d')
now = pd.to_datetime(now,format = '%Y%m%d')
cust['Recent_Date'] = pd.to_datetime(cust.Recent_Date, format='%Y%m%d')
cust.assign(numb_days = now - cust['Recent_Date'] )
```
